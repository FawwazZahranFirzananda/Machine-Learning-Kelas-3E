1. Buatlah model klasifikasi dengan menggunakan SVM untuk data suara, voice.csv.

```python
# Import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Load data
df = pd.read_csv('data/voice.csv')
df.head()
```

```python
X = df.drop('label', axis = 1)
y = df['label']

# Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)

# Buat dan Latih Model SVM
svm_model = SVC(kernel = 'linear')
svm_model.fit(X_train, y_train)

# Evaluasi Model
y_pred = svm_model.predict(X_test)

# Hitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f'accuracy: {accuracy}')

# Print laporan klasifikasi
print(classification_report(y_test, y_pred))

# Print matrix kebingungan
confusion = confusion_matrix(y_test, y_pred)
print('Confusion Matrix: ')
print(confusion)
```

2. Buatlah model klasfikasi Multinomial Naive Bayes dengan ketentuan,
    1. Menggunakan data spam.csv
    2. Fitur CountVectorizer dengan mengaktifkan stop_words
    3. Evaluasi hasilnya

```python
# Load data menggunakan encoding
df = pd.read_csv('data/spam.csv', encoding='latinâ€“1') 
df.head()
```

```python
# Drop kolom
df = df.drop(df.iloc[:, 2:], axis=1)
df.head()
```

```python
# Rename kolom
new_cols = {
    'v1': 'Labels',
    'v2': 'SMS'
}

df = df.rename(columns = new_cols)
df.head()
```

```python
# Memisahkan fitur dan label
X = df['SMS'].values
y = df['Labels'].values

# Inisiasi CountVectorizer
cv = CountVectorizer(stop_words = 'english')

# Fitting dan transform X_train dengan CountVectorizer
X = cv.fit_transform(X)

# Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)

# Inisiasi MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train, y_train) # Fit model

# Print hasil
print('training accuracy: ', mnb.score(X_train, y_train))
print('test accuracy: ', mnb.score(X_test, y_test))
```
