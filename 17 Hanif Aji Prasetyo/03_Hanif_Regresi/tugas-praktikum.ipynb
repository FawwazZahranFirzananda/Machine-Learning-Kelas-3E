```python
# Mengimpor library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Mengimpor dataset
dataset = pd.read_csv('data/insurance.csv')

# 1. Identifikasi variabel-variabel yang akan digunakan sebagai variabel bebas (fitur) dan variabel target (biaya medis personal)
X = dataset[['age','bmi']] # Age dan BMI merupakan variabel bebas (fitur), karena berupa data numerik 
y = dataset['charges'] # Charges merupakan variabel target (biaya medis personal)

# 2. Bagi dataset menjadi data latih (train) dan data uji (test) dengan proporsi yang sesuai.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
# 3. Buat model multiple linear regression menggunakan Scikit-Learn
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# 4. Latih model pada data latih dan lakukan prediksi pada data uji.
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# 5. Evaluasi model dengan menghitung metrik seperti R-squared, MSE, dan MAE.
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# R-Squared dan MSE
r_squared = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Hasil evaluasi
print('R-Squared : ', r_squared)
print('MSE       : ', mse)
print('MAE       : ', mae)

# Visualisasi data uji dan prediksi
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.title('Data Aktual vs. Hasil Prediksi')
plt.xlabel('Data Aktual (Charges)')
plt.ylabel('Hasil Prediksi (Charges)')
plt.grid(True)
plt.show()
```

#### Poin utama dari hasil ini adalah bahwa model Linear Regression yang telah dilatih memiliki R2 sebesar 0.10676050395208692, menunjukkan bahwa model yang telah dilatih mampu menjelaskan sekitar 11% dari variasi dalam data testing, yang berarti model ini cukup baik dalam menjelaskan variasi dalam data, meskipun masih ada variasi yang tidak dapat dijelaskan oleh model
