```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2

# Load Gambar
img = cv2.imread ('/51-E-3083-PAH-10-21_jpeg.rf.427345355203f5c87224af98868891d1.jpg')

# Tampilkan gambar
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

h, w, ch = 100, 100, 3

def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data

    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T

    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=14)

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)

fig, ax = plt.subplots(1, 2, figsize=(16, 6),
                       subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace=0.05)
# Load your original image using cv2.imread instead of "img" if you have one
# img = cv2.imread('your_image.jpg')
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title('Original Image', size=16)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat data dengan garis tetap lurus dan titik-titik kecil
data = np.linspace(0, 1, 10000).reshape(10000, 1)
data = np.hstack((data, data, data))  # Salin komponen merah ke hijau dan biru
noise = np.random.normal(0, 0.05, (10000, 3))  # Tambahkan noise ke data
data += noise

# Menentukan fungsi pembengkokan yang mempertahankan garis tetap lurus
def make_curve(data):
    x, y, z = data.T
    x = x  # Tetapkan x sama
    y = y - 0.1  # Mendorong garis sedikit ke bawah
    return np.column_stack((x, y, z))

data = make_curve(data)

# Definisikan atau impor fungsi plot_pixels di sini
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data

    # Pilih subset acak dari data
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T

    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    ax[0].scatter(R, G, color=colors, marker='.', s=2)  # Perkecil ukuran titik
    ax[0].set(xlabel='Black', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.', s=2)  # Perkecil ukuran titik
    ax[1].set(xlabel='Black', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=14)

# Memanggil fungsi plot_pixels dengan warna yang berbeda, garis yang tetap lurus tapi lebih kecil, dan titik-titik yang sangat kecil
colors = np.random.rand(10000, 3)  # Membuat warna acak
plot_pixels(data, title='Input color space: Line with Smaller and Straight Line with Tiny Dots', colors=colors)

# Menampilkan plot
plt.show()
```
