```python
# Mengimpor library
import numpy as np
mengimport library NumPy dengan alias np. 
import matplotlib.pyplot as plt
ntuk mengimport modul pyplot dari library Matplotlib dengan alias plt. 
import pandas as pd
mengimport library Pandas dengan alias pd.
 
# Mengimpor dataset (Pastikan Anda memiliki file CSV 'Posisi_gaji.csv' dalam direktori yang sama)
dataset = pd.read_csv('Posisi_gaji.csv')
yang digunakan untuk membaca dataset dari file CSV dengan nama "Posisi_gaji.csv" menggunakan Pandas. 
X = dataset.iloc[:, 1:2].values
digunakan untuk memilih kolom kedua dari DataFrame (indeks kolom 1) sebagai fitur (variabel independen).
y = dataset.iloc[:, 2].values  # Ubah menjadi satu kolom saja
perintah yang digunakan untuk memilih kolom ketiga dari DataFrame (indeks kolom 2) sebagai target (variabel dependen). 
 
# Feature Scaling
from sklearn.preprocessing import StandardScaler
yang digunakan untuk mengimpor kelas StandardScaler dari modul preprocessing dalam scikit-learn.
sc_X = StandardScaler()
membuat objek StandardScaler untuk melakukan penskalaan pada fitur (X). 
sc_y = StandardScaler()
membuat objek StandardScaler untuk melakukan penskalaan pada target (y). 
X = sc_X.fit_transform(X.reshape(-1, 1))
melakukan penskalaan pada fitur X. 
y = sc_y.fit_transform(y.reshape(-1, 1))
melakukan penskalaan pada target y dengan langkah yang sama seperti yang dilakukan pada fitur X.
 
# Fitting SVR ke dataset
from sklearn.svm import SVR
mengimpor kelas SVR dari modul svm dalam scikit-learn.
regressor = SVR(kernel='rbf')
membuat objek regresor SVM dengan kernel Radial Basis Function ("rbf").
regressor.fit(X, y)
digunakan untuk melatih model SVR menggunakan data yang telah di-skala (X dan y) sebelumnya.
 
# Visualisasi hasil SVR (resolusi tinggi dan kurva yang lebih halus)
X_grid = np.arange(min(X), max(X), 0.01).reshape(-1, 1)
digunakan untuk membuat serangkaian nilai X secara berurutan dalam interval yang ditentukan.
plt.scatter(X, y, color='red')
untuk membuat scatter plot data aktual. Data fitur X dan target y diplot sebagai titik-titik merah di grafik.
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
membuat plot garis prediksi dari model SVR.
plt.title('(SVR)')
memberikan judul kepada plot.
plt.xlabel('Tingkat posisi')
memberikan label sumbu x pada plot.
plt.ylabel('Gaji')
memberikan label sumbu y pada plot.
plt.show()
menampilkan plot yang telah dibuat.
 
# Prediksi hasil
# Buat array 2D yang berisi tingkat posisi yang akan diprediksi
tingkat_posisi_prediksi = np.array([[6.5]])
membuat sebuah array NumPy yang berisi tingkat posisi yang ingin Anda prediksi gajinya.
# Penskalaan fitur untuk data yang akan diprediksi
tingkat_posisi_prediksi = sc_X.transform(tingkat_posisi_prediksi)
menggunakan objek sc_X (StandardScaler yang telah digunakan sebelumnya) untuk melakukan penskalaan pada tingkat posisi yang akan diprediksi.
# Melakukan prediksi menggunakan model SVR
gaji_prediksi = regressor.predict(tingkat_posisi_prediksi)
yang digunakan untuk melakukan prediksi gaji berdasarkan tingkat posisi yang telah di-skala.
# Kembalikan hasil prediksi ke skala aslinya
gaji_prediksi = sc_y.inverse_transform(gaji_prediksi.reshape(-1, 1))
digunakan untuk mengembalikan gaji yang telah di-skala ke dalam skala aslinya.
print("Prediksi Gaji untuk Tingkat Posisi 6.5:", gaji_prediksi[0])
yang digunakan untuk mencetak hasil prediksi gaji yang telah diubah skala kembali untuk tingkat posisi 6.5. Hasil prediksi ini akan dicetak di layar.

# Evaluasi model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
menghitung Mean Absolute Error (MAE) untuk menghitung Mean Squared Error (MSE), r2_score: Ini adalah fungsi untuk menghitung koefisien determinasi (R-squared)
# Membalikkan penskalaan pada data target yang sudah diprediksi
y_actual = sc_y.inverse_transform(y)
menggunakan objek sc_y (StandardScaler yang digunakan sebelumnya untuk target) untuk mengembalikan target yang telah di-skala ke dalam skala aslinya.
y_pred = regressor.predict(X)
digunakan untuk melakukan prediksi pada data pelatihan yang telah di-skala (X) menggunakan model SVR (regressor).

# Menghitung MAE
mae = mean_absolute_error(y_actual, y_pred)
digunakan untuk menghitung MAE (Mean Absolute Error) antara y_actual (nilai target yang sebenarnya) dan y_pred (nilai prediksi) yang telah dihasilkan oleh model.
# Menghitung MSE
mse = mean_squared_error(y_actual, y_pred)
digunakan untuk menghitung MSE (Mean Squared Error) antara y_actual (nilai target yang sebenarnya) dan y_pred (nilai prediksi) yang telah dihasilkan oleh model.

# Menghitung RMSE
rmse = np.sqrt(mse)
menghitung akar kuadrat dari MSE menggunakan fungsi np.sqrt dari NumPy.

# Menghitung R-squared
r2 = r2_score(y_actual, y_pred)
menghitung R-squared antara nilai target yang sebenarnya (y_actual) dan nilai prediksi (y_pred).

print("MAE:", mae)
mencetak nilai MAE (Mean Absolute Error) ke layar. MAE mengukur kesalahan rata-rata absolut antara prediksi dan nilai aktual.
print("MSE:", mse)
mencetak nilai MSE (Mean Squared Error) ke layar.
print("RMSE:", rmse)
mencetak nilai RMSE (Root Mean Squared Error) ke layar.
print("R-squared:", r2)
mencetak nilai koefisien determinasi (R-squared) ke layar. 
```
