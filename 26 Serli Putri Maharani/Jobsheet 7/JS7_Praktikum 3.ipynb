```python
from minisom import MiniSom # impor kelas minisom 
import numpy as np # impor numpy
import matplotlib.pyplot as plt # impor matplotlib 
from sklearn import datasets # impor dataset
```

```python
# Load dataset
iris = datasets.load_iris()
data = iris.data
```

```python
# Normalisasi data
data = data / data.max(axis=0)
```

```python
# Inisialisasi SOM
map_size = (10, 10)
som = MiniSom(map_size[0], map_size[1], data.shape[1], sigma=0.5, learning_rate=0.5)
```

```python
# Inisialisasi bobot secara acak
som.random_weights_init(data)
```

```python
# Pelatihan SOM
num_epochs = 100
som.train_random(data, num_epochs)
```

```python
# Visualisasi hasil SOM
plt.figure(figsize=(8, 8)) # membuat sebuah gambar yang digunakan untuk menampilkan peta SOM
for i, x in enumerate(data): # melakukan iterasi melalui data yang akan dipetakan ke dalam peta SOM
    w = som.winner(x)  # Pemenang untuk sampel x
    plt.text(w[0]+.5, w[1]+.5, str(i+1), color='k', fontdict={'weight': 'bold', 'size': 11}) # menampilkan nomor indeks data pada lokasi pemenang pada peta SOM
plt.pcolor(som.distance_map().T, cmap='bone_r', alpha=.2) # membuat peta jarak SOM yang dihasilkan oleh objek SOM
plt.colorbar() # menambahkan sebuah warna bar
 
plt.show() # menampilkan gambar yang telah dibuat 
```
