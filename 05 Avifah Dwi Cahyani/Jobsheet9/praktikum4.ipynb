### **Pra Pengolahan Data**
Langkah 1 - Import Library

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

Langkah 2 - Load Data

```python
dataset = pd.read_csv('data/Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values #Memilih fitur dari dataset ari kolom indeks ke-3 hingga kolom sebelum kolom terakhir. 
y = dataset.iloc[:, -1].values #Memilih target atau label kelas dari dataset
```

Cek data (X)

```python
print(X)
```

Langkah 3 - Encoding Data Kategorikal

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
#mengubah kolom kedua dari data X (indeks 2 dalam representasi Python yang dimulai dari 0) menjadi bentuk numerik. 
```

```python
print(X)
```

Langkah 4 - Encoding Kolom "Geography" dengan One Hot Encoder

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

Kode ini melakukan transformasi One-Hot Encoding pada kolom kedua dari data X untuk mengubah variabel kategorikal menjadi bentuk biner yang dapat digunakan oleh algoritma pembelajaran mesin. Hasil transformasi ini akan menggantikan kolom kedua dalam data.

```python
print(X)
```

Langkah 5 - Split Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Kode ini membagi dataset menjadi dua subset: data latih (80% dari dataset) dan data uji (20% dari dataset), yang akan digunakan untuk melatih dan menguji model. Hal ini memungkinkan untuk melakukan evaluasi kinerja model pada data yang belum pernah dilihat sebelumnya.

Langkah 6 - Scaling Fitur

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Kode ini melakukan penskalaan fitur menggunakan StandardScaler untuk memastikan bahwa setiap fitur memiliki distribusi dengan rata-rata 0 dan deviasi standar 1.

### **Membuat Model ANN**
Langkah 1 - Inisiasi Model ANN

```python
ann = tf.keras.models.Sequential()
#Membuat objek jaringan saraf tiruan dengan menggunakan model Sequential. 
```

Langkah 2 - Membuat Input Layer dan Hidden Layer Pertama

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

Langkah 3 - Membuat Hidden Layer Kedua

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

Kode ini menambahkan lapisan Dense dengan 6 neuron menggunakan fungsi aktivasi ReLU ke dalam jaringan saraf tiruan. 

Langkah 4 - Membuat Output Layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

Kode ini menambahkan lapisan Dense terakhir dengan satu neuron dan menggunakan fungsi aktivasi sigmoid ke dalam jaringan saraf tiruan. Lapisan ini akan menghasilkan probabilitas kelas positif pada tugas klasifikasi biner.

### **Training Model**
Langkah 1 - Compile Model (Menyatukan Arsitektur) ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

Kode ini menentukan bagaimana jaringan saraf tiruan akan melakukan proses pelatihan. Ini termasuk pemilihan optimizer, fungsi loss, dan metrik evaluasi yang akan digunakan untuk memantau kinerja model.

Langkah 2 - Fitting Model

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

Kode ini melakukan proses pelatihan pada jaringan saraf tiruan dengan menggunakan data latih. Proses ini akan berlangsung selama 100 epoha, di mana pada setiap epoha, data akan dibagi menjadi batch-batch kecil sebesar 32 sampel untuk memperbarui bobot jaringan

### **Membuat Prediksi**
Diberikan informasi sebagai berikut,
- Geography: France
- Credit Score: 600
- Gender: Male
- Age: 40 years old
- Tenure: 3 years
- Balance: $ 60000
- Number of Products: 2
- Does this customer have a credit card ? Yes
- Is this customer an Active Member: Yes
- Estimated Salary: $ 50000

Apakah customer tersebut perlu dipertahankan?

Modelkan Data Baru dan Buat Prediksi

```python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5)
#Memuat satu data baru yang akan diprediksi
```

Hasil prediksi adalah True jika model memprediksi kemungkinan pelanggan akan meninggalkan bank lebih dari 50%, dan False jika sebaliknya.

Prediksi Dengan Data Testing

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

Kode ini melakukan prediksi pada data uji menggunakan model jaringan saraf tiruan, dan kemudian membandingkan hasil prediksi dengan label kelas sebenarnya. Hasilnya adalah array dua kolom, di mana kolom pertama adalah prediksi model dan kolom kedua adalah label kelas sebenarnya.

Cek Akurasi dan Confusion Matrix

Kode ini digunakan untuk mengevaluasi kinerja model jaringan saraf tiruan menggunakan matriks kebingungannya (confusion matrix) dan akurasi.

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred) #Menghitung matriks kebingungannya dengan membandingkan label kelas sebenarnya 
print(cm) #Mencetak matriks 
accuracy_score(y_test, y_pred) #Menghitung akurasi dari model
```
