## **KMeans**

KMeans adalah satu metode unsupervised learning pada machine learning. Metode ini menentukan jumlah cluster sesuai dengan jumlah  yang dipilih. Proses KMeans secara manual, dapat dilihat pada tautan berikut,
Perhitungan Manual KMeans

Pada modul jobsheet ini, kita akan langsung mempraktikkan pembuatan model KMeans dengan menggunakan python. Untuk modul pertama ini, kita akan menggunakan contoh kasus yang sederhana, yaitu dengan menggunakan dataset iris. Sedangkan untuk modul kedua, kita akan melakukan clustering dengan lebih advance, yaitu reduksi warna dengan data gambar

```python
# Persiapan data

# Mengimpor pustaka Pandas ke dalam program yang dimana library ini berfungsi untuk memanipulasi dan analisis data tabular.
import pandas as pd

# Mengimpor pustaka NumPy ke dalam program yang dimana library ini berfungsi untuk komputasi numerik dalam pengolahan data dan array.
import numpy as np

# Mengimpor modul pyplot dari pustaka Matplotlib yang dimana library ini memiliki fungsi-fungsi Matplotlib untuk membuat dan menampilkan plot.
import matplotlib.pyplot as plt

# Mengacu pada subpustaka sklearn yang berfokus pada algoritma klastering (pengelompokan data) dan mengimpor kelas "KMeans" yang dimana berfungsi untuk klastering data.
from sklearn.cluster import KMeans

# Membaca sebuah berkas CSV dengan menggunakan Pandas. Data dari berkas "Iris.csv" akan dimuat pada suatu struktur data yaitu DataFrame (df) yang memungkinkan untuk melakukan manipulasi dan analisis lebih lanjut pada data tersebut.
df = pd.read_csv('Iris.csv')

# Berfungsi untuk mencetak lima baris pertama dari DataFrame (df) lalu menampilkan sampel data dan memeriksa data terlihat, dilanjut dengan fungsi "head()" yang dimana menampilkan kolom-kolom awal dari data.
df.head()
```

```python
# Seleksi Fitur

penjabaran = "Nanti akan muncul tampilan seperti di bawah ini:"
print(penjabaran)

# Membuat variabel X yang akan berisi fitur-fitur dari DataFrame df. Dalam kode ini, df.iloc[:, 1:-1] digunakan untuk memilih kolom-kolom dari DataFrame df mulai dari kolom kedua (indeks 1) hingga kolom sebelum terakhir (indeks -1).
X = df.iloc[:, 1:-1]

# Membuat variabel y yang akan berisi label atau target dari DataFrame df. Dalam kode ini, df.iloc[:, -1] digunakan untuk memilih kolom terakhir dari DataFrame df, yang biasanya berisi label atau target yang ingin diprediksi dalam suatu pemodelan.
y = df.iloc[:, -1]
```

```python
# Plot Data
# Karena data 4 dimensi, maka akan kita coba
# plot cluster berdasarkan Sepal Length dan Sepal Width  saja

penjabaran = "Nanti akan muncul tampilan seperti di bawah ini:"
print(penjabaran)

# Menghasilkan scatter plot yang menggambarkan titik-titik data dari kolom pertama dan kolom kedua dari array atau DataFrame X. 
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], s = 100)
```

```python
# Buat Model KMeans
# Kali ini kita coba menggunakan k=2 - anggap saja kita tidak tahu jumlah label ada 3 :)

from sklearn.cluster import KMeans

# Inisiasi obyek KMeans
cl_kmeans = KMeans(n_clusters=2)

# Fit dan predict model
y_kmeans = cl_kmeans.fit_predict(X)
```

```python
# Plot hasi cluster berdasarkan Sepal Length dan Sepal Width

penjabaran = "Nanti akan muncul tampilan seperti di bawah ini:"
print(penjabaran)

# Membuat scatter plot dengan sumbu x dan y yang mewakili data dari kolom pertama dan kedua DataFrame X. Ukuran titik-titik ditetapkan sebagai 100, dan warna titik-titik diatur berdasarkan hasil klastering yang disimpan dalam y_kmeans. 
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], s = 100, c=y_kmeans)

# Plot centroid
# Mengambil pusat (centroid) dari klaster yang telah ditemukan oleh algoritma K-Means. Variabel centers akan berisi koordinat pusat dari masing-masing klaster dalam data.
centers = cl_kmeans.cluster_centers_

#  Menunjukkan lokasi pusat klaster dengan titik-titik merah pada data yang telah dikelompokkan menggunakan algoritma K-Means.
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
```

```python
# Cek Nilai SSE
print(f'Nilai SSE: {cl_kmeans.inertia_}')
```

```python
# Implementasi Metode Elbow

penjabaran = "Nanti akan muncul tampilan seperti di bawah ini:"
print(penjabaran)

# List nilai SSE
sse = []

# Cari k terbaik dari 1-10
K = range(1,10)

# Cek nilai SSE setiap k
for k in K:
 kmeanModel = KMeans(n_clusters=k)
 kmeanModel.fit(X)
 sse.append(kmeanModel.inertia_)


# Plotting the distortions
# Berfungsi untuk membuat sebuah figur atau gambar dengan ukuran 8x4 (panjang x lebar) unit. 
plt.figure(figsize=(8,4))

# Berfungsi untuk membuat plot garis. K adalah daftar yang berisi nilai K (jumlah kluster) yang diuji, sse adalah daftar yang berisi nilai Sum of Squared Errors (SSE) yang dihitung untuk setiap nilai K, dan "bx-" adalah format plot. "bx-" berarti plot akan menggunakan tanda silang biru ("x") untuk titik data dan garis biru ("-") untuk menghubungkannya.
plt.plot(K, sse, "bx-")

# Berfungsi  untuk memberi label sumbu-X pada plot. Dalam hal ini, label "k" ditempatkan pada sumbu-X, yang mewakili jumlah kluster yang diuji.
plt.xlabel("k")

# Berfungsi untuk memberi label sumbu-Y pada plot. Dalam hal ini, label "SSE" ditempatkan pada sumbu-Y, yang mewakili nilai Sum of Squared Errors.
plt.ylabel("SSE")

# Berfungsi untuk memberi judul plot. 
plt.title("Metode Elbow untuk Mengetahui Jumlah k Terbaik")

# Berfungsi untuk menampilkan plot atau gambar yang telah dibuat sebelumnya.
plt.show()
```

```python
# Cek Nilai SSE setiap k

penjabaran = "Nanti akan muncul tampilan seperti di bawah ini:"
print(penjabaran)

for idx, sse_val in enumerate(sse, start=1):
    print(f'k={idx}; SSE={sse_val}')
```
